```python
#import bibliotek:
import pandas as pd

data = pd.read_html(
    'https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/',
    header=0,
    flavor='html5lib'
)
data
```

```python
# Wybieram pierwszą tabelę ze strony (read_html zwraca listę tabel)
df = data[0]
```

```python
#Zamień nagłówki kolumn na polskie odpowiedniki:
df.rename(columns={
    'TITLE': 'TYTUL',
    'ARTIST': 'ARTYSTA',
    'YEAR': 'ROK',
    'HIGH POSN': 'MAX POZ'
}, inplace=True)
```

```python
#Ilu pojedynczych artystów znajduje się na liście?
df['ARTYSTA'].nunique()
```

```python
#Które zespoły pojawiają się najczęściej na liście?
df['ARTYSTA'].value_counts()
```

```python
#Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami:
df.columns = df.columns.str.title()
```

```python
#Wyrzuć z tabeli kolumnę ‘Max Poz’:
df.drop('Max Poz', axis=1, inplace=True)
```

```python
#W którym roku wyszło najwięcej albumów znajdujących się na liście?:
df['Rok'].value_counts()
```

```python
#Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?
((df['Rok'] >= 1960) & (df['Rok'] <= 1990)).sum()
```

```python
#W którym roku wydany został najmłodszy album na liście?:
df.sort_values('Rok', ascending=False).head(1)
```

```python
#Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście:
df.sort_values('Rok').groupby('Artysta').head(1)
```

```python
# Zapisuję listę najwcześniej wydanych albumów każdego artysty do pliku CSV:
df.sort_values('Rok').groupby('Artysta').head(1).to_csv('najwczesniejsze_albumy.csv', index=False)
```
